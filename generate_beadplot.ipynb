```python
import pandas as pd
import numpy as np
import altair as alt

from collections import OrderedDict
```

```python
distributed_mentoring = pd.read_csv('./output/distributed_mentoring.csv')
distributed_mentoring = distributed_mentoring.drop(['Unnamed: 0', 'acknowledgments', 'references'], axis=1)
distributed_mentoring
```

```python
test = distributed_mentoring.melt().iloc[0:30, :]
test['value']
```

```python
my_dict = {'A': range(2), 'B': range(5), 'C': [float(x) for x in range(7)], 'D': range(4)}
my_dict_df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in my_dict.items() ]))
my_dict_df
```

```python
def fill_nulls(col):
    vals = col[~col.isnull()].values
    vals = np.resize(vals, len(col)) # this will cause the numbers to just repeat, as desired
    return vals
    
my_dict_df = my_dict_df.apply(fill_nulls, axis=0)
my_dict_df
```

```python
def replace_text_and_fill_nulls(col):
    vals = col[~col.isnull()].values
    vals = [i for i in range(len(vals))] # replace with numbers for ease of processing
    vals = np.resize(vals, len(col)) # this will cause the numbers to just repeat, as desired
    return vals

distributed_mentoring_cleaned = distributed_mentoring.apply(replace_text_and_fill_nulls, axis=0)
distributed_mentoring_cleaned
```

```python
long_form_distributed_mentoring = distributed_mentoring_cleaned.melt(var_name='Section', value_name='Paragraph')
long_form_distributed_mentoring
```

```python
alt.Chart(long_form_distributed_mentoring).mark_point(opacity=1).encode(
    x='Paragraph:N',
    y=alt.Y('Section:N', sort=list(OrderedDict.fromkeys(long_form_distributed_mentoring['Section'])))
).properties(
    width=300,
    height=300
)
```
