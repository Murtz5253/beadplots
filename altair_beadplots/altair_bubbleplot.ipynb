```python
import pandas as pd
import numpy as np
import altair as alt
from collections import OrderedDict
```

```python
test_dict = {'Abstract': ['a', 'b', 'c', 'a', 'b', 'c', 'a'], 'Introduction': ['a', 'b', 'c', 'd', 'a', 'b', 'c'], 
             'Related Work': ['a', 'b', 'c', 'd', 'a', 'b', 'c'], 'Methods': ['a', 'b', 'c', 'd', 'e', 'f', 'g'], 
             'Findings': ['a', 'b', 'c', 'd', 'e', 'f', 'a'], 'Discussion': ['a', 'b', 'c', 'd', 'e', 'f', 'g'],
            'Limitation': ['a', 'b', 'a', 'b', 'a', 'b', 'a'], 'Conclusion': ['a', 'b', 'c', 'd', 'a', 'b', 'c']}

test_df = pd.DataFrame(test_dict)
#test_df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in test_dict.items() ]))
```

```python
test_df
```

```python
alt.Chart(test_df).transform_fold(
    list(test_df.columns), as_=['key', 'value']
).mark_point().encode(
    x='value:N',
    y='key:N',
).properties(
    width=200,
    height=300
)
```

# What if I melt the DataFrame first? Then can I specify a size?

```python
long_form_test_df = test_df.melt(var_name='Section', value_name='Paragraph')
long_form_test_df.head()
```

```python
made_up_algorithm_scores = [.2, 1.0, .75, .2, 1.0, .75, .2] + [.3, 1.0, .3, 1.0, .3, 1.0, .3] + [.4, .1, .4, .1, .4, .1, .4] + [.5]*7 + [.6]*7 + [1.0, .8, .4, .3, .4, .8, 1.0] + [.8]*7 + [.1, .2, .3, 1.0, .1, .2, .3]
long_form_test_df['Algorithm Score'] = made_up_algorithm_scores
long_form_test_df
```

```python
alt.Chart(long_form_test_df).mark_point(opacity=1).encode(
    x='Paragraph:N',
    y=alt.Y('Section:N', sort=list(OrderedDict.fromkeys(long_form_test_df['Section']))),
    size='Algorithm Score'
).properties(
    width=200,
    height=300
)
```
